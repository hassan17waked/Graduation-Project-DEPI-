  1.1 üìä Project Objective:
</div>

<p>
This project addresses a common issue in the banking sector: <strong>customer churn</strong>.<br>
Banks face significant revenue loss when customers leave.<br>
Understanding the reasons behind churn and predicting which customers are at risk can help retain them and improve long-term profitability.
</p>

  1.2 üéØ Project Goal:
</div>


<p>The main goals of this project are:</p>
<ul>
  <li>Analyze customer data to identify <strong>churn patterns</strong>.</li>
  <li>Build a machine learning model to predict which customers are likely to churn.</li>
  <li>Provide actionable insights to help the business reduce churn.</li>
</ul>

  1.3 üîç Problem Framing:
</div>

<p>We frame this as a <strong>binary classification</strong> task:</p>
<ul>
  <li><strong>Target variable:</strong> <code>Exited</code> (1 = customer churned, 0 = customer stayed)</li>
  <li><strong>Input features include:</strong>
    <ul>
      <li>Demographics: <code>Age</code>, <code>Gender</code>, <code>Geography</code></li>
      <li>Financials: <code>Balance</code>, <code>EstimatedSalary</code>, <code>CreditScore</code></li>
      <li>Customer Activity: <code>Tenure</code>, <code>IsActiveMember</code>, <code>NumOfProducts</code>, <code>HasCrCard</code></li>
    </ul>
  </li>
</ul>
